```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Rysowanie wykresow

```python
def plot(start, stop, amt, title):
    f = lambda x : x**2 + 5
    x = np.linspace(start, stop, amt)
    y = [f(i) for i in x]

    plt.figure()
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    
plot(-1, 1, 30, 'Wykres 1')
plot(-6, 6, 30, 'Wykres 2')
plot(0, 5, 30, 'Wykres 3')
```
