```python
import pandas as pd

df = pd.read_csv('../dataset/drinks.csv')
df
```

```python
sorted_by_total = df.sort_values('total_litres_of_pure_alcohol', ascending=False).reset_index(drop=True)
sorted_by_total
```

```python
sorted_by_total.to_csv('../results/sorted_by_total.csv')
```

```python
zero_litres_countries = df[df['total_litres_of_pure_alcohol'] == 0]
zero_litres_countries
```

```python
zero_litres_countries.to_csv('../results/zero_litres_countries.csv')
```

```python
sober_filtered_out = df[df['total_litres_of_pure_alcohol'] > 0]
sober_filtered_out

sober_filtered_out.to_csv('../results/sober_filtered_out.csv')
```
